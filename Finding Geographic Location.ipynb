```python
import pandas as pd
import geonamescache as gc


df = pd.read_csv('news_headline_info.csv')
gnc = gc.GeonamesCache()

latitude = []
longitude = []
country_code = []
for index in df.index:
    city = df['cities'][index]
    val = gnc.get_cities_by_name(city)
    
    if (city == "NaN" or val == []):
        latitude.append('NaN')
        longitude.append('NaN')
        country_code.append('NaN')
    elif (len(val) == 1):
        keys = [e for e in val[0]]
        latitude.append(val[0][keys[0]]['latitude'])
        longitude.append(val[0][keys[0]]['longitude'])
        country_code.append(val[0][keys[0]]['countrycode'])
    else:
        maxpop = 0
        index = 0
        for j in range(len(val)):
            keys = [e for e in val[j]]
            population = val[j][keys[0]]['population']
            if (population > maxpop):
                maxpop = population
                index = j
        keys = [e for e in val[index]]
        latitude.append(val[index][keys[0]]['latitude'])
        longitude.append(val[index][keys[0]]['longitude'])
        country_code.append(val[index][keys[0]]['countrycode'])

df['latitude'] = latitude
df['longitude'] = longitude
df['countrycode'] = country_code
del df['countries']

df.to_csv('latitude_longitude.csv')
df
```
